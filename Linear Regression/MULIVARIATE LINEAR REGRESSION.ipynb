```python
import pandas as pd
data = pd.read_csv("housing.csv")
data
```

```python
data_x = data.drop(["median_house_value", "ocean_proximity","total_bedrooms"], axis = 1)
data_x
```

```python
data_y = data.loc[:,["median_house_value"]]
data_y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
data_x_train, data_x_test, data_y_train, data_y_test = train_test_split(data_x, data_y, test_size = 0.2)
```

```python
model.fit(data_x_train, data_y_train)
data_y_predict = model.predict(data_x_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
print("Mean Squared Error: ", mean_squared_error(data_y_test, data_y_predict))
print("Theta_N:",model.coef_)
print("Theta_0:",model.intercept_)

print("R2 Score:",r2_score(data_y_test, data_y_predict))

'''
R2 Score 1 means model has its best fit
R2 Score 0 means model has a poor fit
R2 Score -ve means model has worst line of fit
'''
```
