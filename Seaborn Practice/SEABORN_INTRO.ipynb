```python
import seaborn as sns
data = sns.load_dataset("tips")
data
```

```python
data.corr() 
'''
correltion positive means the two variables are proportional to each other
correlation negative means two variables are inversly proportional to each other
'''
```

```python
sns.heatmap(data.corr())
```

```python
sns.jointplot(x="tip", y = "total_bill", data = data, kind = "hex")
```

```python
sns.jointplot(x="tip", y = "size", data = data, kind = "reg")
```

```python
sns.pairplot(data, hue = "time")
```

```python
data["time"].value_counts()
```

```python
sns.distplot(data["tip"],kde = False, color = "green")
```
