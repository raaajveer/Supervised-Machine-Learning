```python
data["current_year"] = 2021
```

```python
data
```

```python
data["years_driven"] = data["current_year"] - data["year"]
```

```python
data
```

```python
data_y =  data["selling_price"]
```

```python
data_x = data.drop(labels = ["name","year","current_year", "selling_price"], axis = 1)
data_x
```

```python
data_x  = pd.get_dummies(data = data_x, drop_first = True)
data_x
```

```python
from sklearn.linear_model import LinearRegression
linreg_model = LinearRegression()

from sklearn.model_selection import train_test_split
data_x_train, data_x_test, data_y_train, data_y_test = train_test_split(data_x, data_y, test_size = 0.2)
```

```python
linreg_model.fit(data_x_train, data_y_train)
linreg_y_pred = linreg_model.predict(data_x_test)
from sklearn.metrics import r2_score
r2_score(data_y_test, linreg_y_pred)
```

```python
from sklearn.svm import SVR
svr_model = SVR(kernel = "rbf")
svr_model.fit(data_x_train, data_y_train)
svr_y_pred = svr_model.predict(data_x_test)
r2_score(data_y_test, svr_y_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor()
rf_model.fit(data_x_train, data_y_train)
rf_y_pred = rf_model.predict(data_x_test)
r2_score(data_y_test, rf_y_pred)
```

```python
data_pd_compare = {"Predicted": rf_y_pred, "Actual": data_y_test}
data_compare = pd.DataFrame(data = data_pd_compare)
data_compare
```
