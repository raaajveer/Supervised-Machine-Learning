```python
data = data.drop(labels = "bike_name", axis = 1)
```

```python
data_dummied = pd.get_dummies(data = data,drop_first = True )
data_dummied
```

```python
data_dummied.shape
```

```python
data_x = data_dummied.drop(labels = "price", axis = 1)
data_y = data_dummied["price"]
data_x
```

```python
from sklearn.model_selection import train_test_split
data_x_train, data_x_test, data_y_train, data_y_test = train_test_split(data_x, data_y, test_size = 0.1)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(data_x_train, data_y_train)
data_y_pred = model.predict(data_x_test)
```

```python
from sklearn.metrics import r2_score
print(r2_score(data_y_test, data_y_pred) * 100)
```

```python
y_compare_1 = {"Predicted": data_y_pred, "Actual": data_y_test}
y_compare_1
```

```python
y_compared_1 = pd.DataFrame(data = y_compare_1)
y_compared_1
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
plt.scatter(x = y_compare_1["Predicted"], y = y_compare_1["Actual"])
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor(n_estimators = 10)
rf_model.fit(data_x_train, data_y_train)
rf_y_pred = rf_model.predict(data_x_test)
```

```python
r2_score(y_true = data_y_test, y_pred = rf_y_pred)
```

```python
y_compare_2 = {"Actual": data_y_test, "Predicted": rf_y_pred}
y_compare_2 = pd.DataFrame(data = y_compare_2)
y_compare_2
```

```python
#RandomSearchCV
from sklearn.model_selection import RandomizedSearchCV
import numpy as np

#Parameters
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1000, num= 10)]
criterion = ["mse", "mae"]
max_depth = [int(x) for x in np.linspace(start = 10, stop = 1000, num = 10 )]
min_samples_split = [float(x) for x in np.linspace(start = 0.1, stop = 1, num = 5)]
min_samples_leaf = [int(x) for x in np.linspace(start = 1, stop = 10, num = 5)]
max_features = ["auto", "sqrt", "log2"]
    
random_params = {
    "n_estimators" : n_estimators,
    "max_depth" : max_depth,
    "min_samples_split" : min_samples_split,
    "min_samples_leaf" : min_samples_leaf,
    "max_features" : max_features
}
random_params
```

```python
random_cv_model = RandomizedSearchCV(estimator = rf_model, param_distributions = random_params, n_iter = 100, verbose = 2, cv = 3, random_state = 100, n_jobs = -1, error_score = "raise")
random_cv_model.fit(data_x_train, data_y_train)
random_y_pred = random_cv_model.predict(data_x_test)
print("Accuracy of the RandomizedSearchCV Model", r2_score(data_y_test, random_y_pred), "%")
random_cv_model.cv_results_
```

```python
print(random_cv_model.best_params_)
n_estimators = [random_cv_model.best_params_["n_estimators"] - 100, random_cv_model.best_params_["n_estimators"], random_cv_model.best_params_["n_estimators"] + 100]
max_depth = [random_cv_model.best_params_["max_depth"] - 100, random_cv_model.best_params_["max_depth"], random_cv_model.best_params_["max_depth"] + 100]
min_samples_split = [random_cv_model.best_params_["min_samples_split"], random_cv_model.best_params_["min_samples_split"] + 0.1]
min_samples_leaf = [random_cv_model.best_params_["min_samples_leaf"]]
max_features = [random_cv_model.best_params_["max_features"]]

grid_params = {
    "n_estimators" : n_estimators,
    "max_depth" : max_depth,
    "min_samples_split" : min_samples_split,
    "min_samples_leaf" : min_samples_leaf,
    "max_features" : max_features
    
}
grid_params
```

```python
from sklearn.model_selection import GridSearchCV

grid_model = GridSearchCV(estimator= rf_model, param_grid= grid_params, verbose= 2, cv= 3)
grid_model.fit(data_x_train, data_y_train)
grid_y_pred = grid_model.predict(data_x_test)
print("Accuracy of the GridSearchCV Model", r2_score(data_y_test, random_y_pred), "%")
random_cv_model.cv_results_
```
